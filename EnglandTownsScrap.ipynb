```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_towns_in_England"

res = requests.get(url).text
soup = BeautifulSoup(res,'lxml')
```

```python
town_data = []

for tables in soup.find_all('table', class_='wikitable'):
    for items in tables.find_all('tr')[1:]:
        data = items.find_all(['td'])
        try:
            town = data[0].a.text
            county = data[1].text
            
            town_data.append([town, county])
        except IndexError:
            print(data)

        print("{}|{}".format(town,county))
        
towns_df = pd.DataFrame(town_data, columns=["town", "county"])
```

```python
towns_df.to_csv("town_data/town_data.csv", index=False)
```
